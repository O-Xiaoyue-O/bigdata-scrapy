# **使用 Homebrew 安裝**

1.  使用 Homebrew 能輕鬆安裝許多不好裝(設定)的套件，例：ChromeDriver
2.  注意你的空間：Homebrew 安裝程式接近 600MB


安裝 macOS 缺少套件的管理工具 Homebrew: https://brew.sh/index_zh-tw
```
$ /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
```

安裝完成後，再執行以下指令，新增 Homebrew 路徑到 PATH
```
 $  echo 'eval "$(/opt/homebrew/bin/brew shellenv)"' >> ~/.zprofile
 $  eval "$(/opt/homebrew/bin/brew shellenv)"
```

安裝 python3
```
$ brew install python3
Python has been installed as
  /opt/homebrew/bin/python3
```

# ** 移除 Homebrew **
執行以下指令
```
$ /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/uninstall.sh)"
$ rm -rf /opt/homebrew
$ rm ~/.zprofile
```

# 使用 Homebrew 安裝 ChromeDriver 及使用 Python 虛擬環境安裝 Selenium 及 lxml


### 安裝 ChromeDriver

```
$ brew install cask chromedriver
```

### 安裝 pip 虛擬環境

```
$ python3 -m venv myenv
```

### 切換到指定的 pip 虛擬環境，此處為 myenv

```
$ source ~/myenv/bin/activate
```

### 安裝 Selenium

```
$ python3 -m pip install selenium
```

### 安裝lxml

```
$ python3 -m pip install lxml
```

# Visual Studio Code 設定 Python 環境

1. 從上方功能列中「檔案」下拉選單，打開「開啟...」，開啟「工作區」或「資料夾」

2. 從上方功能列中「檢視」下拉選單，打開「終端機」，在終端機內輸入


```
% source ~/myenv/bin/activate
```

3. 再一次從功能列中「檢視」下拉選單，打開「命令選擇區...」，輸入「InterPreter」，即會看到如下圖，點擊 Python: 選擇直譯器

Python選擇直譯器.png

4. 再選擇 ★ Python 3.9.12 64-bit /opt/homebrew/bin/python3 ，如下圖

brew-python3.png

如此，之後的 python 的未來套件安裝與編譯動作，都在 Visual Studio Code 的終端機內執行即可😀

## 從 Python 官網下載及安裝


1.   如果你喜歡手動設置，以調校最佳的作業環境參數，就安裝這個版本
2.   下載路徑：https://www.python.org/downloads/


選擇您想要安裝的 python3 版本，

*   如果 Mac 是 Intel 的 CPU，選擇：macOS 64-bit Intel-only installer
*   如果 Mac 是 M1 系列的 CPU，選擇：macOS 64-bit universal2 installer


程式會安裝在：
```
/Application/python3
/usr/local/bin/python3
```

#移除 Python 安裝

1.  從應用程式中直接移除

uninstall_python3.png

2. 執行以下移除指令

```
$ sudo rm -rf /usr/local/bin/python*
$ sudo rm -rf /usr/local/bin/pip*
```
