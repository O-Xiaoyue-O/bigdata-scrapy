以下環境是將 Colab 的 Python 版本升到 3.9，
又因為 root permission 的原因，必須使用虛擬環境才能將 selenium 裝起來，
但在執行程式時，卻因為沒辦法使用虛擬環境，所以才失敗，
如果能作到執行 Notebook 時，使用的是虛擬環境，以下程式才能夠正確。

```python
#安裝 python 3.9
!apt-get update -y
!apt-get install python3.9

#變更版本替代
!update-alternatives --install /usr/bin/python3 python3 /usr/bin/python3.7 1
!update-alternatives --install /usr/bin/python3 python3 /usr/bin/python3.9 2
!update-alternatives --config python3

#檢查 Python 版本
!python --version
```

```python
#安裝&升級 Python 相關套件
!apt-get install python3.9-distutils
!apt-get install python3-pip
!python -m pip install --upgrade --force-reinstall pip
```

```python
#建立虛擬環境(因為在 colab 沒辦法使用 root 權限)
!apt-get install python3.9-venv
!python3.9 -m venv myEnv
!source myEnv/bin/activate
!python --version
```

```python
#使用虛擬環境的 pip 安裝 selenium
!myEnv/bin/python -m pip install selenium
```

```python
#安裝 Chromedriver
!apt-get install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
```
